```python
import os
import pdfplumber
import pandas as pd
import numpy as np
import re
from collections import defaultdict
import logging
import traceback

logging.basicConfig(filename = 'invoice_processing.log', level=logging.INFO, format='%(asctime)s:%(levelname)s:%(message)s')

class Invoice:
    """
    A class used to represent an invoice. It extracts information from a PDF invoice and stores it in a pandas DataFrame.
    """
    def __init__(self, pdf_path):
        try:
            if not pdf_path.endswith('.pdf'):
                logging.error(f"Invalid file path. Only PDF files are supported. {pdf_path}")
                raise ValueError(f"Invalid file path. Only PDF files are supported. {pdf_path}")
            if not os.path.exists(pdf_path):
                logging.error(f"File not found. Please check the file path. {pdf_path}")
                raise FileNotFoundError(f"File not found. Please check the file path. {pdf_path}")
            self.pdf_path = pdf_path
            self.text = self.extract_text()
            self.order_info = self.extract_order_info()
            self.sku_details_df = self.extract_sku_details()
            self.to_dataframe = self.to_dataframe()
        except Exception as e:
            logging.error(f"Failed to process invoice: {pdf_path}  \n {e}")
            traceback.print_exc()

    def extract_text(self):
        """Extract text from PDF file"""
        text = ""
        try:
            with pdfplumber.open(self.pdf_path) as pdf:
                for page in pdf.pages:
                    text = text + '\n' + page.extract_text()
        except pdfplumber.pdfparser.PDFSyntaxError:
            logging.error(f"Invalid PDF file. Please check the file contents. {self.pdf_path}")
            traceback.print_exc()
            raise ValueError("Invalid PDF file. Please check the file contents.")
        return text

    def extract_order_info(self):
        """Extract general order information from text"""
        order_info = defaultdict(list)

        patterns = {
            'Order Number': r"Invoice [a-zA-Z0-9, ]+",
            'Order Date': r"Order Date: ([a-zA-Z0-9, ]+)",
            'Shipping Address': r"SHIPPING ADDRESS\n([\s\S]+?)\nActive Memberships",
            'Membership Type': r"Active Memberships when order placed: ([a-zA-Z0-9, ]+)",
            'Email': r"✉ ([\w\.-]+@[\w\.-]+)",
            'Phone': r"\uf095 ([\d-]+)",
            'Subtotal': r"Subtotal: \$([^\n\r]*)",
            'Discount': r"Discount: -\$([^\n\r]*)",
            'Shipping': r"Shipping: ([^\n\r]*)",
            'Payment Method': r"Payment method: ([a-zA-Z ]+)",
            'Total': r"Total: \$([^\n\r]*)"
        }

        for field, pattern in patterns.items():
            try:
                match = re.search(pattern, self.text)
                if match:
                    order_info[field] = match.group(1) if match.groups() else match.group(0)
                else:
                    logging.warning(f"{field} not found. Adding null value.")
                    print(f"{field} not found. Adding null value.")
                    order_info[field] = None
            except Exception as e:
                logging.error(f"Failed to extract {field} from invoice. {e}")
                traceback.print_exc()
                raise ValueError(f"Failed to extract {field} from invoice.")
        try:
            sections = self.text.split('\n')
            coupons = [section[2:] for section in sections if section.startswith('\uf02b')]
            order_info['Coupon'] = coupons
        except:
            logging.warning("Coupon not found. Adding null value.")
            print("Coupon not found. Adding null value.")
            order_info['Coupon'] = None

        # Create DataFrame for general order information
        general_order_df = pd.DataFrame([order_info])
        general_order_df['Shipping Address'] = general_order_df['Shipping Address'].str.replace('\n', ', ')
        # # Regular expression patterns for extracting specific fields
        # order_number_pattern = r"Invoice [a-zA-Z0-9, ]+"
        # order_date_pattern = r"Order Date: ([a-zA-Z0-9, ]+)"
        # shipping_address_pattern = r"SHIPPING ADDRESS\n([\s\S]+?)\nActive Memberships"
        # membership_type_pattern = r"Active Memberships when order placed: ([a-zA-Z0-9, ]+)"
        # email_pattern = r"✉ ([\w\.-]+@[\w\.-]+)"
        # phone_pattern = r"\uf095 ([\d-]+)"
        # subtotal_pattern = r"Subtotal: \$([^\n\r]*)"
        # discount_pattern = r"Discount: -\$([^\n\r]*)"
        # shipping_pattern = r"Shipping: ([^\n\r]*)"
        # payment_method_pattern = r"Payment method: ([a-zA-Z ]+)"
        # total_pattern = r"Total: \$([^\n\r]*)"

        # try: 
        #     order_info['Order Number'].append(re.search(order_number_pattern, self.text).group(0))
        # except:
        #     print("Order Number not found. Adding null value.")
        #     order_info['Order Number'].append(None)
        # try:
        #     order_info['Order Date'].append(re.search(order_date_pattern, self.text).group(1))
        # except:
        #     print("Order Date not found. Adding null value.")
        #     order_info['Order Date'].append(None)
        # try:
        #     order_info['Shipping Address'].append(re.search(shipping_address_pattern, self.text).group(1).replace('\n', ', '))
        # except:
        #     print("Shipping Address not found. Adding null value.")
        #     order_info['Shipping Address'].append(None)
        # try:
        #     order_info['Membership Type'].append(re.search(membership_type_pattern, self.text).group(1))
        # except:
        #     print("Membership Type not found. Adding null value.")
        #     order_info['Membership Type'].append(None)
        # try:
        #     order_info['Email'].append(re.search(email_pattern, self.text).group(1))
        # except:
        #     print("Email not found. Adding null value.")
        #     order_info['Email'].append(None)
        # try:
        #     order_info['Phone'].append(re.search(phone_pattern, self.text).group(1))
        # except:
        #     print("Phone not found. Adding null value.")
        #     order_info['Phone'].append(None)
        # try:
        #     order_info['Subtotal'].append(re.search(subtotal_pattern, self.text).group(1))
        # except:
        #     print("Subtotal not found. Adding null value.")
        #     order_info['Subtotal'].append(None)
        # try:
        #     order_info['Discount'].append(re.search(discount_pattern, self.text).group(1))
        # except:
        #     print("Discount not found. Adding null value.")
        #     order_info['Discount'].append(None)
        # try:
        #     order_info['Shipping'].append(re.search(shipping_pattern, self.text).group(1))
        # except:
        #     print("Shipping not found. Adding null value.")
        #     order_info['Shipping'].append(None)
        # try:
        #     order_info['Payment Method'].append(re.search(payment_method_pattern, self.text).group(1))
        # except:
        #     print("Payment Method not found. Adding null value.")
        #     order_info['Payment Method'].append(None)
        # try:
        #     order_info['Total'].append(re.search(total_pattern, self.text).group(1))
        # except:
        #     print("Total not found. Adding null value.")
        #     order_info['Total'].append(None)
        # try:
        #     sections = first_page_text.split('\n')
        #     coupons = [section[2:] for section in sections if section.startswith('\uf02b')]
        #     order_info['Coupon'].append(coupons)
        # except:
        #     print("Coupon not found. Adding null value.")
        #     order_info['Coupon'].append(None)

        # # Create DataFrame for general order information
        # general_order_df = pd.DataFrame({
        #     'Order Number': [order_info['Order Number'][0]],
        #     'Order Date': [order_info['Order Date'][0]],
        #     'Shipping Address': [order_info['Shipping Address'][0]],
        #     'Membership Type': [order_info['Membership Type'][0]],
        #     'Email': [order_info['Email'][0]],
        #     'Phone': [order_info['Phone'][0]],
        #     'Subtotal': [order_info['Subtotal'][0]],
        #     'Discount': [order_info['Discount'][0]],
        #     'Shipping': [order_info['Shipping'][0]],
        #     'Payment Method': [order_info['Payment Method'][0]],
        #     'Total': [order_info['Total'][0]],
        #     'Coupon': [order_info['Coupon'][0]]
        # })
        return general_order_df

    def extract_sku_details(self):
        """Extract SKU and product details from text"""
        order_info = defaultdict(list)

        # Split the text into sections based on known delimiters
        sections = self.text.split('\n')
        
        try:
            # Find the index where SKU and product details start and end
            start_idx = sections.index('SKU Product Quantity Price') + 1
            end_idx = len(sections)
        except ValueError:
            print("The invoice does not contain SKU and product details.")

        # Extract the SKU and product details from the relevant section
        sku_details_text = '\n'.join(sections[start_idx:end_idx])

        # Initialize lists to store product information
        sku_names = []
        product_names = []
        quantities = []
        prices = []
        product_types = []

        # Split the SKU details text into lines
        lines = sku_details_text.split('\n')

        # Initialize variables to hold the current product type and other details
        current_product_type = None

        # Loop through each line to extract SKU and product details
        for line in lines:
            if 'DOSAGE' in line or 'Subtotal:' in line or 'Discount:' in line or 'Shipping:' in line or 'Payment method:' in line or 'Total:' in line or 'coupon used:' in line:
                continue  # Skip lines that don't contain SKU and product details
            elif re.match(r'^[A-Z][a-zA-Z ]+$', line):  # Check if the line is a product type
                current_product_type = line
            else:
                parts = line.split()
                if len(parts) >= 4:  # Check if the line contains enough parts to be valid SKU and product details
                    sku_names.append(parts[0])
                    quantities.append(parts[-2])
                    prices.append(parts[-1])
                    product_names.append(" ".join(parts[1:-2]))
                    product_types.append(current_product_type)

        # Add SKU and product details to the order information dictionary
        order_info['SKU Name'] = sku_names
        order_info['Product Name'] = product_names
        order_info['Quantity'] = quantities
        order_info['Price'] = prices
        order_info['Product Type'] = product_types

        # Create DataFrame for SKU and product details
        sku_details_df = pd.DataFrame({
            'SKU Name': order_info['SKU Name'],
            'Product Name': order_info['Product Name'],
            'Quantity': order_info['Quantity'],
            'Price': order_info['Price'],
            'Product Type': order_info['Product Type']
        })

        # first convert the quantity column to numeric else null values will be added
        sku_details_df['Quantity'] = pd.to_numeric(sku_details_df['Quantity'], errors='coerce')
        # drop the rows with null quantity
        sku_details_df.dropna(subset=['Quantity'], inplace=True)
        sku_details_df['Quantity'] = sku_details_df['Quantity'].astype(int)
        sku_details_df['Product Type'] = sku_details_df['Product Type'].replace('SKU Product Quantity Price', np.nan)
        # Forward-fill NaN values in 'Product Type'
        sku_details_df['Product Type'].ffill(inplace=True)
        # reset the index
        sku_details_df.reset_index(drop=True, inplace=True)
        return sku_details_df

    def to_dataframe(self):
        """Combine general order information and SKU and product details into a single DataFrame"""
        # Repeat order_info to match the number of rows in sku_details_df
        order_info_repeated = pd.concat([self.order_info] * len(self.sku_details_df), ignore_index=True)

        # Concatenate the repeated order_info with sku_details_df along axis 1
        result = pd.concat([order_info_repeated, self.sku_details_df], axis=1)
        return result
```

```python
import imaplib
import email
import os
from email.header import decode_header
import traceback
import logging

logging.basicConfig(filename = 'email_attachment_extraction.log', level=logging.INFO, format='%(asctime)s:%(levelname)s:%(message)s')

class EmailAttachmentExtractor:
    """
    Class for connecting to an email inbox via IMAP4 and extracting attachments from emails.
    
    Attributes
    ----------
    email : str
        Email address of the account to connect to.
    password : str
        Password of the account to connect to.
    imap_server : str
        IMAP server to connect to.

    Methods
    -------
    connect()
        Connect to the IMAP server and log in.
    close_connection()
        Close the connection to the IMAP server.
    extract_pdf_attachments(directory_to_save, num_emails=5)
        Extract PDF attachments from the latest emails that have 'pdf' in their subject.
    """
    def __init__(self, email, password, imap_server = "imap.gmail.com"):
        self.email = email
        self.password = password
        self.imap_server = imap_server

    def connect(self):
        """
        Connect to the IMAP server and log in.
        
        Returns
        -------
        bool
            True if the connection was successful, False otherwise.
        """
        try:
            self.mail = imaplib.IMAP4_SSL(self.imap_server)
            self.mail.login(self.email, self.password)
            logging.info("Successfully logged in to email account %s", self.email)
            return True
        except imaplib.IMAP4.error as e:
            logging.error("Login failed: %s", e)
            print(f"Login failed: {e}")
            return False

    def close_connection(self):
        """
        Close the connection to the IMAP server.
        
        Returns
        -------
        None
        """
        self.mail.close()
        self.mail.logout()
        logging.info("Successfully logged out from email account %s", self.email)

    def extract_pdf_attachments(self, directory_to_save=None, num_emails=5, subject_contains='pdf', date_from=None, date_to=None):
        """
        Extract PDF attachments from the latest emails that have subject_contains in their subject and were received between date_from and date_to.
        
        Parameters
        ----------
        directory_to_save : str
            Directory where the attachments will be saved.
        num_emails : int
            Number of emails to process.
        subject_contains : str
            Substring to search for in the email subject.
        date_from : str
            Date in the format 'dd-mm-yyyy' to search for emails received after this date.
        date_to : str
            Date in the format 'dd-mm-yyyy' to search for emails received before this date.
        """
        try:
            # Select the mailbox to search in; 'INBOX' by default
            self.mail.select("INBOX")

            # Search for emails with pdf_contains in their subject in the mailbox and were received between date_from and date_to
            criteria = []
            if subject_contains:
                criteria.append(f'SUBJECT "{subject_contains}"')
            if date_from:
                criteria.append(f'SINCE {date_from}')
            if date_to:
                criteria.append(f'BEFORE {date_to}')
            search_criteria = ' '.join(criteria) if criteria else 'ALL'

            status, messages = self.mail.search(None, search_criteria)
            if status != "OK":
                logging.error("Failed to search emails: %s", messages)
                print(f"Failed to search emails: {messages}")
                return

            # The search returns a list of email IDs; we'll get the latest 'num_emails' emails
            messages = messages[0].split(b' ')
            if not messages or messages == [b'']:
                logging.warning("No emails found with subject containing '%s'", subject_contains)
                print(f"No emails found with subject containing '{subject_contains}'")
                return
            
            latest_emails = messages[-num_emails:]

            pdfs = []

            for mail in latest_emails:
                # Fetch the email by ID
                status, data = self.mail.fetch(mail, "(RFC822)")
                for response_part in data:
                    if isinstance(response_part, tuple):
                        # Parse the email content
                        msg = email.message_from_bytes(response_part[1])
                        subject, encoding = decode_header(msg["Subject"])[0]
                        if isinstance(subject, bytes):
                            subject = subject.decode(encoding)
                        print(f"Processing email subject: {subject}")
                        logging.info("Processing email subject: %s", subject)

                        # Check if the email has any attachments
                        for part in msg.walk():
                            if part.get_content_maintype() == "multipart" or part.get("Content-Disposition") is None:
                                continue
                            if "attachment" in part.get("Content-Disposition"):
                                file_name = part.get_filename()
                                if file_name and file_name.endswith(".pdf"):
                                    binary_data = part.get_payload(decode=True)
                                    pdf = {
                                        "file_name": file_name,
                                        "binary_data": binary_data
                                    }
                                    pdfs.append(pdf)
                                    print(f"Found PDF attachment: {file_name}")
                                    logging.info("Found PDF attachment: %s", file_name)
                                    if directory_to_save:
                                        file_path = os.path.join(directory_to_save, file_name)
                                        with open(file_path, "wb") as f:
                                            f.write(binary_data)
                                        print(f"Saved attachment: {file_path}")
                                        logging.info("Saved attachment: %s", file_path)
            return pdfs

        except Exception as e:
            print(f"An error occurred: {e}")
            logging.error("An error occurred: %s", e)
            traceback.print_exc()
        finally:
            self.close_connection()
```
